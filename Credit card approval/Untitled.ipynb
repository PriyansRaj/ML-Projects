```python
X =df.drop(columns=['Ind_ID','Phone','EMAIL_ID','Work_Phone','Mobile_phone','Birthday_count','label','Car_Owner'],axis=1)
```

```python
y= df['label']
```

```python
X
```

```python
num_col = X.select_dtypes(exclude='object').columns
cat_col = X.select_dtypes(include='object').columns
num_col
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
scaler = StandardScaler()
ohe = OneHotEncoder(handle_unknown='ignore')
transformer = ColumnTransformer( [
        ("OneHotEncoder",ohe,cat_col),
        ("StandardScaler",scaler,num_col)
    ]
                               )
    
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train = transformer.fit_transform(X_train)
```

```python
X_test = transformer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
```

```python
models = {
    'logistic Regression':LogisticRegression(),
    'Support Vector Classifier':SVC(),
    'KNN': KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Gradient Boost" : GradientBoostingClassifier(),
    "Random Forest":RandomForestClassifier(),
    "AdaBoost" : AdaBoostClassifier(),
    "Xgboost": XGBClassifier()
}
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,f1_score,precision_score
def metric_Score(y_test,y_pred):
    cm = confusion_matrix(y_test,y_pred)
    acc = accuracy_score(y_test,y_pred)
    Recall = recall_score(y_test,y_pred)
    score= f1_score(y_test,y_pred)
    prec = precision_score(y_test,y_pred)
    return cm,acc,Recall,score,prec
    
```

```python
for i in range(len(models)):
    model  = list(models.values())[i]
    model.fit(X_train,y_train)
    name = list(models.keys())[i]
    
    print(f"\n\nMetrics scores for {name}")
    print("\nCross-validation scores")
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print("Cross-validation scores for each fold:", cv_scores)
    print("Average cross-validation score:", cv_scores.mean())
    print("\n For the train data")
    y_pred = model.predict(X_train)
    cm,acc,recall,score,prec = metric_Score(y_train,y_pred)
    print("Confusion Matrix",cm)
    print("Accuracy score",acc)
    print("Recall score",recall)
    print("F1-Score",score)
    print("Precison",prec)

    print("For the test data")
    y_pred = model.predict(X_test)
    cm,acc,recall,score,prec = metric_Score(y_test,y_pred)
    print("Confusion Matrix",cm)
    print("Accuracy score",acc)
    print("Recall score",recall)
    print("F1-Score",score)
    print("Precison",prec)
    
    
```

```python
from sklearn.model_selection import GridSearchCV
param_grids = {
    'logistic Regression': {
        'C': [0.1, 1, 10],
        'solver': ['liblinear', 'saga'],
        'max_iter': [100, 200]
    },
    'Support Vector Classifier': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf','poly'],
        'gamma': ['scale', 'auto']
    },
    'KNN': {
        'n_neighbors': [3, 5, 7, 10,12,15],
        'weights': ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
    },
    'Decision Tree': {
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'Gradient Boost': {
        'n_estimators': [50, 100, 150],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 4, 5],
        'subsample': [0.8, 1.0]
    },
    'Random Forest': {
        'n_estimators': [100, 200, 300],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'bootstrap': [True, False]
    },
    'AdaBoost': {
        'n_estimators': [50, 100, 150],
        'learning_rate': [0.01, 0.1, 1]
    },
    'Xgboost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 6, 9],
        'subsample': [0.8, 1.0]
    }
}
```

```python
best_models = {}

for name, model in models.items():
    print(f"Performing grid search for {name}...")
    
    
    grid_search = GridSearchCV(estimator=model, param_grid=param_grids[name], cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
    
    # Fit the grid search
    grid_search.fit(X_train, y_train)
    
    # Get the best parameters and model
    best_models[name] = grid_search.best_estimator_
    
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best score for {name}: {grid_search.best_score_}")
```

```python
best_models
```

```python

model_scores = {}


for name, model in best_models.items():
    print(f"Evaluating {name} on test data...")
    
    y_pred = model.predict(X_test)
    
    cm, acc, recall, score, prec = metric_Score(y_test, y_pred)
    
    model_scores[name] = {
        'accuracy': acc,
        'recall': recall,
        'f1_score': score,
        'precision': prec,
        'confusion_matrix': cm
    }


for name, scores in model_scores.items():
    print(f"\n{name} metrics:")
    print(f"Accuracy: {scores['accuracy']}")
    print(f"Recall: {scores['recall']}")
    print(f"F1-Score: {scores['f1_score']}")
    print(f"Precision: {scores['precision']}")
    print(f"Confusion Matrix: \n{scores['confusion_matrix']}")


best_model_name = max(model_scores, key=lambda x: model_scores[x]['f1_score']) 
print(f"\nThe best model based on F1-Score is: {best_model_name}")
```

```python
best_model = best_models[best_model_name]
```

```python
import pickle
pickle.dump(best_model,open("KNN.pkl",'wb'))
```
