```python
df_new.replace({'Sex':{"male":0,"female":1},"Embarked":{"S":0,"C":1,'Q':2}},inplace=True)
```

```python
X = df_new.drop(columns=['Name','Ticket','Survived'],axis=1)
```

```python
y= df_new['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(X_train_scaled,y_train)
```

```python
y_predict = logistic.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
score= accuracy_score(y_test,y_predict)
matrix = confusion_matrix(y_test,y_predict)
```

```python
score
```

```python
matrix
```

```python
sns.heatmap(matrix,annot=True)
```

```python

from sklearn.svm import SVC
```

```python
svc= SVC()
```

```python
svc.fit(X_train_scaled,y_train)
```

```python
y_pred = svc.predict(X_test_scaled)
```

```python
score= accuracy_score(y_test,y_pred)
matrix = confusion_matrix(y_test,y_pred)
```

```python
score
```

```python
sns.heatmap(matrix,annot=True)
```

```python
from sklearn.metrics import roc_curve,auc
```

```python
probabilities = logistic.predict_proba(X_test_scaled)[:, 1]
```

```python
fpr,tpr, thresholds = roc_curve(y_test,probabilities)
```

```python
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],'b--')
plt.xlabel("False Positive Rate - FPR")
plt.ylabel("True Positive Rate - TPR ")
plt.title("Receiver Operating Characteristics - ROC Curve")
plt.text(0.6,0.5,"Baseline")
plt.text(0.3,0.8,"ROC Curve")
```

```python
from sklearn.metrics import roc_auc_score
 
roc_auc_score(y_test, probabilities)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(logistic,X_train_scaled,y_train,cv=5,scoring='roc_auc').mean()
```

```python
cross_val_score(svc,X_train_scaled,y_train,cv=5,scoring='roc_auc').mean()
```

```python
param_grid = {
    'C': [0.1, 1, 10],  # Regularization strength
    'solver': ['liblinear', 'saga']  # Solver for optimization
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=logistic, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)
```

```python
print("Best Hyperparameters: ", grid_search.best_params_)

# Get the best model
best_model = grid_search.best_estimator_

# Evaluate on the test set
test_accuracy = best_model.score(X_test, y_test)
print(f"Test Set Accuracy: {test_accuracy}")
```

```python
y_predict = best_model.predict(X_test_scaled)
```

```python
score= accuracy_score(y_test,y_predict)
matrix = confusion_matrix(y_test,y_predict)
```
