```python
X = zoo_df.drop(['Class_Type','Class_Number','animal_name'],axis=1)
y = zoo_df['Class_Type']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
scaler =StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(multi_class='multinomial', solver='lbfgs'),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": XGBClassifier(objective='multi:softprob', num_class=len(np.unique(y_train_encoded)))
}
    
```

```python
for name,model in models.items():
    model.fit(X_train_scaled,y_train_encoded)
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test_encoded,y_pred)
    acc_score.append(acc)
    print(f"Accuracy score for {name} is {acc}")
    print(f"Error rate for {name} is {np.mean(y_pred!=y_test_encoded)}")
    print()
```

```python
from scipy.stats import randint, uniform
from sklearn.model_selection import RandomizedSearchCV
param_grids = {
    "Logistic Regression": {
        'C': uniform(0.1, 10),
        'penalty': ['l2', 'none'],
        'solver': ['lbfgs', 'saga'],
    },
    'Naive Bayes':{'var_smoothing': uniform(1e-9, 1e-1)},
    "KNN": {
        'n_neighbors': randint(1, 20),
        'weights': ['uniform', 'distance'],
        'metric': ['minkowski', 'euclidean', 'manhattan'],
    },
    "Decision Tree": {
        'max_depth': randint(1, 20),
        'min_samples_split': randint(2, 10),
        'min_samples_leaf': randint(1, 10),
        'criterion': ['gini', 'entropy'],
    },
    "Random Forest": {
        'n_estimators': randint(50, 200),
        'max_depth': randint(1, 20),
        'min_samples_split': randint(2, 10),
        'min_samples_leaf': randint(1, 10),
        'bootstrap': [True, False],
    },
    "Gradient Boosting": {
        'n_estimators': randint(50, 200),
        'learning_rate': uniform(0.01, 0.2),
        'max_depth': randint(1, 10),
        'subsample': uniform(0.5, 1),
        'loss': ['log_loss', 'exponential'],
    },
    "AdaBoost": {
        'n_estimators': randint(50, 200),
        'learning_rate': uniform(0.01, 1),
    },
}
models = {
    "Logistic Regression": LogisticRegression(multi_class='multinomial', solver='lbfgs'),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
}
    
```

```python
results = {}
for name, model in models.items():
    print(f"Running RandomizedSearchCV for {name}...")
    randomsearch = RandomizedSearchCV(estimator= model,n_iter=10,cv=5,verbose=2,n_jobs=-1,param_distributions=param_grids[name],random_state=42)
    randomsearch.fit(X_train_scaled,y_train_encoded)
    results[name] = randomsearch.best_params_,randomsearch.best_score_
    
```

```python
for model_name, (best_params, best_score) in results.items():
    print(f"Best parameters for {model_name}: {best_params}")
    print(f"Best cross-validation score for {model_name}: {best_score:.4f}")
    print()
```
